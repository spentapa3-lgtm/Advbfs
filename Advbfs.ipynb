```python
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['G', 'H'],
    'E': ['I', 'J'],
    'F': [],
    'G': ['K', 'L'],
    'H': [],
    'I': [],
    'J': [],
    'K': [],
    'L': []
}

source = 'A'
destination = 'L'

visited = set()
non_visited = set(graph.keys())
path = []

def print_state(step, current):
    print(f"\n----- STEP {step} -----")
    print("Source Node :", source)
    print("Current Node :", current)
    print("Children     :", graph[current])
    print("Visited Nodes:", visited)
    print("Non-Visited  :", non_visited)
    print("Path         :", " -> ".join(path))

current = source
visited.add(current)
non_visited.remove(current)
path.append(current)

print_state(1, current)

selected = 'D'
visited.add(selected)
non_visited.remove(selected)
path.append(selected)

print("\nAfter Selecting:", selected)
print("Visited Nodes:", visited)
print("Non-Visited  :", non_visited)
print("Path         :", " -> ".join(path))

current = selected

print_state(2, current)

selected = 'G'
visited.add(selected)
non_visited.remove(selected)
path.append(selected)

print("\nAfter Selecting:", selected)
print("Visited Nodes:", visited)
print("Non-Visited  :", non_visited)
print("Path         :", " -> ".join(path))

current = selected

print_state(3, current)

selected = destination
visited.add(selected)
non_visited.remove(selected)
path.append(selected)

print("\nAfter Selecting:", selected)
print("Visited Nodes:", visited)
print("Non-Visited  :", non_visited)
print("Path         :", " -> ".join(path))

print("\n----- FINAL RESULT -----")
print("Shortest Path from A to L:")
print(" -> ".join(path))
```
